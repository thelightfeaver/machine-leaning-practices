```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
df_costumer = pd.read_csv('../datasets/mall_customers.csv')
df_costumer.head()
```

```python
# rename columns
df_costumer.rename(columns={'CustomerID': 'customer_id', 
                            'Gender': 'gender', 
                            'Age': 'age', 
                            'Annual Income (k$)': 'annual_income', 
                            'Spending Score (1-100)': 'spending_score'}, inplace=True)
df_costumer.columns.to_list()
```

```python
df_gender = df_costumer["gender"].value_counts()
plt.bar(df_gender.index, df_gender.values, color=['pink', 'steelblue'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')
plt.show()
```

```python
df_mean_spending= df_costumer.groupby("gender")["spending_score"].agg("mean")
df_mean_spending

plt.pie(df_mean_spending.values, labels=df_mean_spending.index, autopct='%1.1f%%', colors=['pink', 'steelblue'])
plt.title('Average Spending Score by Gender')
plt.show()
```
